# Stats

## Missing value

```python
missing_value_handler_funcs = {
    "raw": functools.partial(
        lipana.do_missing_value_handling_on_df,
        config=None,
        filter_less_than_rep=2,
    ),
    "FillFullEmpty": functools.partial(
        lipana.do_missing_value_handling_on_df,
        config=lipana.FullEmptyFillingMissingValueHandler(min_rep_count=2),
        filter_less_than_rep=2,
    ),
    "FillFullEmpty_NormalDistSample": functools.partial(
        lipana.do_missing_value_handling_on_df,
        config=(
            lipana.FullEmptyFillingMissingValueHandler(min_rep_count=2),
            lipana.NormalDistSamplingMissingValueHandler(
                min_rep_count=2, do_imputation_for_rep_range=(1, 1), log_scale_minmax_diff=1
            ),
        ),
        filter_less_than_rep=2,
    ),
}
```

## FC

```python
fc_configs = {
    "mean_div": lipana.RatioCalcConfig(
        cond_to_cols_map=exp_setting.condition_to_runs_map,
        base_cond="H5_Y100",
        cond_pairs=None,
        is_log=True,
        temp_reverse_log_scale=2,
        div_method="agg_and_divide",
        agg_method="mean",
        new_colname_pattern="log2_fc",
    )
}
```

## Stats

```python
hypo_test_configs = {
    "limma_cut_site": lipana.LimmaConfig(
        annotation_data=exp_setting,
        entry_name="cut_site",
        condition_pairs="H5_Y100",
        output_column_map="default",
        output_column_map_override={"logFC": "log2_fc_limma", "adj.P.Val": "adj_pvalue_limma"},
        recollected_columns="log2_fc",
        dump_in_df_to=None,
        del_files=True,
    ),
    "limma_precursor": lipana.LimmaConfig(
        annotation_data=exp_setting,
        entry_name="precursor",
        condition_pairs="H5_Y100",
        output_column_map="default",
        output_column_map_override={"logFC": "log2_fc_limma", "adj.P.Val": "adj_pvalue_limma"},
        recollected_columns="log2_fc",
        dump_in_df_to=None,
        del_files=True,
    ),
}
```

## Sign

```python
sign_voting_configs = {
    "t_sign_on_cut_site": lipana.SignVotingConfig(
        sign_value_col="t",
        group_col="cut_site",
        row_sign_col="row_sign",
        group_sign_col="group_sign",
        drop_opposite_sign=True,
        drop_bidirectional_balanced_group=True,
    ),
    "fc2p05_on_cut_site": lipana.SignVotingConfig(
        pos_sign_gt_col_value=(("log2_fc", np.log2(2)),),
        pos_sign_lt_col_value=(("pvalue", 0.05),),
        neg_sign_gt_col_value=None,
        neg_sign_lt_col_value=(("log2_fc", np.log2(1 / 2)), ("pvalue", 0.05)),
        group_col="cut_site",
        row_sign_col="row_sign",
        group_sign_col="group_sign",
        filter_value_col="pvalue",
    ),
    "fc1d5p05_on_cut_site": lipana.SignVotingConfig(
        pos_sign_gt_col_value=(("log2_fc", np.log2(1.5)),),
        pos_sign_lt_col_value=(("pvalue", 0.05),),
        neg_sign_gt_col_value=None,
        neg_sign_lt_col_value=(("log2_fc", np.log2(1 / 1.5)), ("pvalue", 0.05)),
        group_col="cut_site",
        row_sign_col="row_sign",
        group_sign_col="group_sign",
        filter_value_col="pvalue",
    ),
}


stats_agg_configs = {
    "fc_limmafc_p_adjp_cut_site": lipana.ValueAggregationConfig(
        group_col="cut_site",
        agg_col=("log2_fc", "log2_fc_limma", "pvalue", "adj_pvalue_limma"),
        agg_func=(
            np.nanmedian,
            np.nanmedian,
            lipana.scipy_comb_p_with_nan,
            lipana.scipy_comb_p_with_nan,
        ),
        recollected_columns=None,
    ),
}
```

## FDR

```python
fdr_configs = {
    "exp_level": lipana.FDRConfig(
        group=None,
        pre_filter=None,
        p_col="pvalue",
        new_col_name="adj_pvalue_exp_level",
        method="BH",
    ),
    "protein_level": lipana.FDRConfig(
        group="protein_group",
        pre_filter=None,
        p_col="pvalue",
        new_col_name="adj_pvalue_protein_level",
        method="BH",
    ),
}
```

## On site

```python
site_qdf
```

```python
funcs = [
    functools.partial(
        missing_value_handler_funcs["FillFullEmpty"],
        exp_runs=exp_setting.condition_to_runs_map["H7d5_Y100"],
        ctrl_runs=exp_setting.condition_to_runs_map["H5_Y100"],
        annotation_col="cut_site",
    ),
    functools.partial(
        lipana.do_desc_summary_on_df,
        config=fc_configs["mean_div"],
    ),
    functools.partial(
        lipana.do_stats_on_df,
        exp_runs=exp_setting.condition_to_runs_map["H7d5_Y100"],
        ctrl_runs=exp_setting.condition_to_runs_map["H5_Y100"],
        config=hypo_test_configs["limma_cut_site"],
    ),
    functools.partial(
        lipana.attach_annotation_from_other_df,
        other_df=report.id_report,
        annotation_cols="protein_group",
        on="cut_site",
        method="leftjoin",
    ),
    functools.partial(
        lipana.do_stats_on_df,
        config=list(fdr_configs.values()),
    ),
]
```

```python
site_stats = site_qdf.clone()
for func in funcs:
    site_stats = func(site_stats)
```

```python
funcs = [
    functools.partial(
        missing_value_handler_funcs["FillFullEmpty"],
        exp_runs=exp_setting.condition_to_runs_map["H7d5_Y100"],
        ctrl_runs=exp_setting.condition_to_runs_map["H5_Y100"],
        annotation_col="precursor",
    ),
    functools.partial(
        lipana.do_desc_summary_on_df,
        config=fc_configs["mean_div"],
    ),
    functools.partial(
        lipana.do_stats_on_df,
        exp_runs=exp_setting.condition_to_runs_map["H7d5_Y100"],
        ctrl_runs=exp_setting.condition_to_runs_map["H5_Y100"],
        config=hypo_test_configs["limma_precursor"],
    ),
    functools.partial(
        lipana.attach_annotation_from_other_df,
        other_df=report.id_report,
        annotation_cols="cut_site",
        on="precursor",
        pre_filter=(~pl.col("cut_site_is_restricted")),
        method="leftjoin",
    ),
    functools.partial(
        lipana.do_stats_on_df,
        config=[
            sign_voting_configs["t_sign_on_cut_site"],
            stats_agg_configs["fc_limmafc_p_adjp_cut_site"],
        ],
    ),
    functools.partial(
        lipana.attach_annotation_from_other_df,
        other_df=report.id_report,
        annotation_cols="protein_group",
        on="cut_site",
        method="leftjoin",
    ),
    functools.partial(
        lipana.do_stats_on_df,
        config=list(fdr_configs.values()),
    ),
]
```

```python
site_stats_from_prec = prec_qdf.clone()
for func in funcs:
    site_stats_from_prec = func(site_stats_from_prec)
```
